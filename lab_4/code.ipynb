```python
divider = 0

try:
    print("Спроба виконати ділення...")
    result = 10 / divider
    print(f"Результат: {result}")
except ZeroDivisionError as e:
    print(f"Виникла помилка: {e} (не можна ділити на нуль)")
except Exception as e:
    print(f"Сталася непередбачена помилка: {e}")
finally:
    print("Блок обробки помилок завершив роботу.")
```

```python
def a_b_func(a, b):
    return a, b

# Створення лямбда-функції
my_lambda = lambda name, age: f'Студент: {name}, Вік: {age}'

print("Результат звичайної функції:", a_b_func("Тест", 100))
print("Результат лямбди:", my_lambda("Віталій", 19))

# Виклик лямбди 
print("Комбінований виклик:", my_lambda(*a_b_func("Олексій", 20)))
```

Відповідь АІ: Python у Jupyter Notebook — це потужний інструмент, де код виконується в окремих комірках. Основні концепції, які варто засвоїти на початку:

Динамічна типізація: Вам не потрібно вказувати тип змінної (int, str), Python визначить його автоматично.

Списки (Lists): Це впорядковані колекції, які можуть містити різні типи даних.

Функції: Використовуються для групування коду, який виконує певну задачу.

```python
# 1. Змінні та типи даних
name = "Віталій"
age = 19
is_student = True

print(f"Привіт! Мене звати {name}, мені {age} років. Статус студента: {is_student}")

# 2. Робота зі списками
# Створимо список технологій та додамо нову
technologies = ["Python", "Jupyter", "Git"]
technologies.append("VS Code")

print("\nСписок технологій, які я вивчаю:")
for tech in technologies:
    print(f"- {tech}")

# 3. Функція для розрахунку
def calculate_discount(price, discount_percent):
    """Функція повертає ціну зі знижкою"""
    final_price = price - (price * discount_percent / 100)
    return final_price

original_price = 1000
sale_price = calculate_discount(original_price, 15)

print(f"\nПочаткова ціна: {original_price} грн.")
print(f"Ціна зі знижкою 15%: {sale_price} грн.")
```
